 第二节 使用多个图形和坐标轴
 子窗口应用

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)#e^-t*cos(2*pi*t)

t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.1)#从零到5 每隔0.1记一次数，不包括0.5 同range()函数的意义

plt.figure(1) #第一幅大图。它包含两个子窗
plt.subplot(211)#subplot()对应（numrows,numcols,fignum），即2行1列第1幅图
plt.plot(t1,f(t1),'bo',t2,f(t2),'g',lw=2.0)#两条曲线

plt.subplot(212)
plt.plot(t2, np.sin(2*np.pi*t2),'ro')

plt.show()

```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current 一定要制定是在那幅图上加title
plt.title('Easy as 1, 2, 3') # subplot 211 title
plt.show()
```

进阶 - 另一种子窗口表示

```python
import matplotlib.pyplot as plt
import numpy as np 

#create some data to use for the plot
dt = 0.001
t = np.arange(0.,10.,dt)
x = np.random.randn(len(t))
r = np.exp(-t[:1000]/0.05)        #impulse response 列表切片表示
s = np.convolve(x,r)[:len(x)]*dt  #colored noise

#the main axes is subplot111 by default
plt.plot(t,s)
plt.axis([0,1,1.1*np.amin(s),2*np.amax(s)])
plt.ylabel('current (nA)')
plt.xlabel('time (s)')
plt.title('Gaussian colored noise')

#this is an instet  axes over the main axes
a = plt.axes([.6, .6, .2, .2], facecolor='y')
n, bins, patches = plt.hist(s, 400, normed=1)
plt.title('Probability')
#plt.xticks([])
#plt.yticks([])

# this is another inset axes over the main axes
a = plt.axes([0.2, 0.6, .2, .2], facecolor='y')
plt.plot(t[:len(r)], r)
plt.title('Impulse response')
plt.xlim(0, 0.2)
#plt.xticks([])
#plt.yticks([])

plt.show()
```

第三节 working with text  添加坐标标签、图表名称和文字

```python
import numpy as np 
import matplotlib.pyplot as plt

#fix random state for reproducibility(再现性)
np.random.seed(19680801)

mu,sigma = 100,15

x= mu+sigma*np.random.randn(10000)

#the histogram of the data
n, bins, patches = plt.hist(x,50,normed=1,facecolor='g',alpha=0.45)# x后的数即bins意思是把x分几份。
#alpha后面是透明度的意思，不透明是1。normed=1 猜测是正太分布的意思

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.text(60,.025,r'$\mu=100,\ \sigma=15$')
#注意\mu,和\sigma 是字符表示。中间\+空格，是为了加空格。 60，0.25是起始位置坐标
#字符串之前的r，表示该字符串是一组原始字符串，而不是将反斜杠作为python的转义处理
plt.axis([40,160,0,0.03])# x轴和y轴范围
plt.grid(True) #显示内部网格线
plt.show()
```

对数坐标轴和其他非线性坐标轴--plt.xscale('log')

```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval [0, 1]
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
